##### Dealing with outliers for horsepower

```python
# IQR
Q1 = np.percentile(auto['horsepower'], 25,
                   interpolation = 'midpoint')
 
Q3 = np.percentile(auto['horsepower'], 75,
                   interpolation = 'midpoint')
IQR = Q3 - Q1
print(IQR)
```

```python
# Above Upper bound
upper = auto["horsepower"] >= (Q3+1.5*IQR)
 
#print("Upper bound:",upper)
print(np.where(upper))
 
# Below Lower bound
lower = auto['horsepower'] <= (Q1-1.5*IQR)
#print("Lower bound:", lower)
print(np.where(lower))
```

```python
for x in ['horsepower']:
    q75,q25 = np.percentile(auto.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    auto.loc[auto[x] < min,x] = np.nan
    auto.loc[auto[x] > max,x] = np.nan
```

### Feature Scaling:

```python
from sklearn.preprocessing import StandardScaler
scaled_features = StandardScaler().fit_transform(df.values)
df = pd.DataFrame(scaled_features, index=df.index, columns=df.columns)
df.head()
```

### Splitting of data into Training and Testing Data
Now that we have explored the data a bit, itâ€™s time to go ahead and split our initial data into training and testing subsets.
Here we set a variable X i.e, independent columns as the numerical features of the dataframe ,and a variable y i.e, dependent column as the
<b>â€œpriceâ€ </b> column

```python
X=df[['curb_weight', 'wheel_base','length','width','engine_size','bore', 'stroke','horsepower','city_mpg', 'highway_mpg']]
y=df[['price']]
```

```python
X
```

```python
y
```

#### Use model_selection.train_test_split from sklearn to split the data into training and testing sets. Set test_size=0.20 and random_state=105

1. create test-train split
- 80%: training data
- 20%: test data
<br> <b>this split is for model without regularization hyper-parameters;<br> with regularization, the split would be</b>
- 60%: train data
- 20%: hyper-parameter/regularization validation data
- 20%: test data

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=105) # use 80:20 law for spliting 
```

##### Training the Model using Linear Regression
- Now, at this step we are able to train our model on our training data using Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train,y_train)
```

```python
print("R-square : {}".format(lr_model.score(X_test,y_test))) #no regularization
```

```python
lr_model.coef_
```

#### Predictions on Test Data
- Now that we have train our model, letâ€™s evaluate its performance by doing the predictions on the unseen data

```python
predictions=lr_model.predict(X_test)
print("predictions:", lr_model.predict(X_test))
```

```python
sns.set(rc={"figure.figsize":(8,4)})
plt.scatter(y_test,lr_model.predict(X_test))
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
plt.show()
```

```python
#regularizing the linear model
from sklearn.linear_model import Ridge
ridge_reg_1=Ridge(alpha=1,normalize=True)
ridge_reg_1.fit(X_train,y_train)
ridge_reg_1.score(X_test,y_test)   #alpha =1
```

```python
#regularizing the linear model

ridge_reg_05=Ridge(alpha=0.5,normalize=True)
ridge_reg_05.fit(X_train,y_train)
ridge_reg_05.score(X_test,y_test)   #alpha =0.5
```

```python
#regularizing the linear model
ridge_reg_2=Ridge(alpha=2,normalize=True)
ridge_reg_2.fit(X_train,y_train)
ridge_reg_2.score(X_test,y_test)   #alpha =2
```

#### Evaluating the Model
- To evaluate our model performance, we will be calculating the residual sum of squares and the explained variance score (R2)

```python
### Determine the metrics such as Mean Absolute Error, Mean Squared Error, and the Root Mean Squared Error.

from sklearn import metrics
print('MAE :'," ", metrics.mean_absolute_error(y_test,predictions))
print('MSE :'," ", metrics.mean_squared_error(y_test,predictions))
print('RMAE :'," ", np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

#### Explore the Residuals
- By observed the metrics calculated in the above steps, we should have a very good model with a good fit. Now, letâ€™s quickly explore the residuals to make sure that everything was okay with our dataset and finalize our model.

###### To see the above thing, try to plot a histogram of the residuals and make sure it looks normally distributed

```python
plt.figure(figsize=(15,8))
sns.distplot(y_test - predictions,bins=50)
plt.xlabel('Residual')
plt.show()
```
